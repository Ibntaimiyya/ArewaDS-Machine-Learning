Lets begin with position salaries

```python
import pandas as pd
pos_df = pd.read_csv('Position_salaries.csv')
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

X = pos_df.iloc[:, 1:2].values
y = pos_df.iloc[:, 2].values

lin_reg = LinearRegression()
lin_reg.fit(X, y)

poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

y_pred = lin_reg.predict(X)
y_pred_poly = lin_reg_2.predict(X_poly)

print("R^2 score (Linear Regression):", r2_score(y, y_pred))
print("R^2 score (Polynomial Regression):", r2_score(y, y_pred_poly))

plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='blue')
plt.plot(X, y_pred_poly, color='green')
plt.title('Truth or Bluff (Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

1. Its clear from the above visualization that Polynomial Regression is the best fit for prediction of Position salaries for workers.
2. The performance R^2 score of Polynomial Regression was : 0.9973922891706614 (99.7%) which is better than the performance of Linear Regression: 0.6690412331929895; approximately (66.9%)

NOw lets see our ice_cream_sales data


```python
ice_sales = pd.read_csv('Ice_Cream_Sales.csv')
X_ice = ice_sales.iloc[:, :-1].values
y_ice = ice_sales.iloc[:, -1].values

lin_reg_ice = LinearRegression()
lin_reg_ice.fit(X_ice, y_ice)

poly_ice = PolynomialFeatures(degree=4)
X_ice_poly = poly_ice.fit_transform(X_ice)
lin_reg_ice_2 = LinearRegression()
lin_reg_ice_2.fit(X_ice_poly, y_ice)

y_ice_pred = lin_reg_ice.predict(X_ice)
y_ice_pred_poly = lin_reg_ice_2.predict(X_ice_poly)

print("R^2 score (Linear Regression):", r2_score(y_ice, y_ice_pred))
print("R^2 score (Polynomial Regression):", r2_score(y_ice, y_ice_pred_poly))

plt.scatter(X_ice, y_ice, color='red')
plt.plot(X_ice, y_ice_pred, color='blue')
plt.plot(X_ice, y_ice_pred_poly, color='green')
plt.title('Ice Cream Sales Prediction (Regression)')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

1. Also for the ice cream sales data; The Polynomial Regression is a better fit than Linear Regression.
2. The score of the Polynomial Regression was : 99.9% accurate and that of the LinearRegression was 99.07%
