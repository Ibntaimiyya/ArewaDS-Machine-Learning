```python
#Filter out the required bird to work on i.e Snow goose
Snow_goose = birds[birds['Name']=='Snow goose']
Snow_goose
```

```python
# Length range
x = np.linspace(Snow_goose['MinLength'], Snow_goose['MaxLength'])

# Body mass range
y = np.linspace(Snow_goose['MinBodyMass'], Snow_goose['MaxBodyMass'])

# Wing span range
z = np.linspace(Snow_goose['MinWingspan'], Snow_goose['MaxWingspan'])
```

```python
#Line Plot
# Body Mass against Length
plt.plot(x,y,color='red')
plt.title('Length vs Body Mass')
plt.xlabel("Length (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

```python
# Wing Span against Length
plt.plot(x,z,color='red')
plt.title('Length vs Wing span')
plt.xlabel("Length (cm)")
plt.ylabel("Wing span (cm)")
plt.grid(alpha=0.4)
plt.show()
```

```python
# Wing Span and Body Mass
plt.plot(z, y,color='red')
plt.title('Body Mass vs Wing Span')
plt.xlabel("Wing Span (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

The general conclusion about the snow geese, belonging to the family of anatide of anser genus and order anserformes of ducks/waterfowl category, is that. It's length, mass and wing span are propotional to each other. The larger it's length the heavier it will be and the longer will be it's wing span.
