```python
import pandas as pd
import matplotlib.pyplot as plt
anime_pet = pd.read_csv('pets.csv')
print(anime_pet)

```

```python
# first: some month were written in 'abbreviated' form while some were written in 'full' form. So we need to convert them all to the same format.  
# lets convert all the months to 'full' form' by mapping categories to fewer ones.
# create a dictionary to map abbreviated months to full months
month_map = {
    'JAN': 'January', 'january': 'January',
    
    'Sept': 'September'
}
# replace abbreviated months with full months
anime_pet['birth_month'] = anime_pet['birth_month'].replace(month_map)
# now we can convert the 'month' column to a categorical type with ordered categories
print(anime_pet['birth_month'].unique())
```

```python
# similarly the state column has some states written in abbreviated form and some in full form. for example CA and California, FL and Florida, etc.
# so lets convert them to full form as well. CA to California, FL to Florida, AL to Alabama, etc.
state_map = {
    'CA': 'California', 'FL': 'Florida', 'AL': 'Alabama', 'RI': 'Rhode Island', 'AK': 'Alaska',
}

# replace abbreviated states with full states
anime_pet['state'] = anime_pet['state'].replace(state_map)
print(anime_pet['state'].unique())
print(anime_pet)

```

```python
# Their is a missing value in the 'state' column ('NaN') we need to replace it
anime_pet['state'] = anime_pet['state'].fillna('New York')# assuming 'New York' is the default state for missing values
# let convert the plural 'cats' to singular 'cat' in the 'pet' column
anime_pet['pet'] = anime_pet['pet'].replace({'Cats': 'Cat'})
# and also we need to convert the 'birth_month' column to a categorical type with ordered categories
anime_pet['birth_month'] = pd.Categorical(anime_pet['birth_month'], 
                                            categories=['January', 'April', 'July', 'September', 'November', 'December'], 
                                            ordered=True)
print(anime_pet)
```

```python
# we can now begin Data Visualization
# lets plot the number of pets by state using a bar chart using matplotlib
plt.figure(figsize=(10, 6))
anime_pet['state'].value_counts().plot(kind='bar')
plt.title('Number of Pets by State')
plt.xlabel('State')
plt.ylabel('Number of Pets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting count of pets by birth month using bar chart with matplotlib
plt.figure(figsize=(10, 6))
anime_pet['birth_month'].value_counts().plot(kind='bar')
plt.title('Count of Pets by Birth Month')
plt.xlabel('Birth Month')
plt.ylabel('Count of Pets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# finally, lets plot the number of pets by type using a bar chart with matplotlib
plt.figure(figsize=(10, 6))
anime_pet['pet'].value_counts().plot(kind='bar')
plt.title('Number of Pets by Type')
plt.xlabel('Pet Type')
plt.ylabel('Number of Pets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
